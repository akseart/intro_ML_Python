# Задача 1. И снова скобочные последовательности
Создать $n$ правильных скобочных последовательностей длины $2n$

````python
"()()()"
"(())()"
"((()))"
def create_bracket_seq(n):
    res = []
    for j in range(n):
       res.append("()" * j + "(" * (n - j) + ")" * (n - j)  )
    return res

create_bracket_seq(3)
````

````python
def get_human_readable_config(config, server_names, result="text.txt"):
    serv_name = {}
    with open(server_names, 'r') as names:
        for i in names:
            name = i.split()
            res = ""
            for i in name[1:]:
                res = res + i + ' '
            serv_name[name[0]] = res

    with open(config, 'r') as conf, open(result, 'w') as res:
        for i in conf:
            ip = i.split()[1]
            res.write(i.rstrip() + " # " + serv_name[ip])
            print(i.rstrip() + " # " + serv_name[ip])

get_human_readable_config("config.txt", "servers.txt")
````

# Задача 5. Свободная касса

Известно, что завтра в магазин придет $n$ посетителей для каждого из них известно точное время прихода. Известно, что касса обслуживает каждого клиента менее чем за минуту. Сколько нужно открыть касс, чтобы каждый посетитель прошел её без очереди?

````python
time_visitors = [
    (8, 0),
    (8, 8),
    (8, 10),
    (8, 10),
    (8, 11),
    (8, 45),
    (8, 45),
    (8, 45),
]
from collections import Counter
def get_num_cashier(times):
    return max(Counter(times).values())


get_num_cashier(time_visitors)
````

````python
def get_num_cashier(times):
    res = {}
    for i in times:
        if i in res:
            res[i] += 1
        else:
            res[i] = 1
    return max(res.values())
get_num_cashier(time_visitors)
````

# Задание 6. Простые числа
Найти все простые числа от $1$ до $N$

````python
%%time
def get_prime(n):
    res = []
    for number in range(n):
        num_divider = 0
        for divider in range(1, number+1):
            if number % divider == 0:
                num_divider += 1
        if num_divider == 2:
            res.append(number)
    return res

get_prime(10000)
````

````python
%%time
def get_prime(n):
    res = []
    for number in range(2, n):
        num_divider = 0
        for divider in range(2, number//2):
            if number % divider == 0:
                num_divider += 1
        if num_divider == 2:
            res.append(number)
    return res

get_prime(10000)
````

Используем "Решето Эратосфена"

1. Выписать подряд все целые числа от двух до n (2, 3, 4, …, n).
2. Пусть переменная p изначально равна двум — первому простому числу.
3. Зачеркнуть в списке числа от 2p до n считая шагами по p (это будут числа кратные p: 2p, 3p, 4p, …).
4. Найти первое незачёркнутое число в списке, большее чем p, и присвоить значению переменной p это число.
5. Повторять шаги 3 и 4, пока возможно.



![](New_Animation_Sieve_of_Eratosthenes.gif)

````python
%%time
def get_prime(n):
    sieve = list(range(n+1))
    sieve[1] = 0
    for p in sieve:
        if p > 1:
            for j in range(2*p, n+1, p):
                sieve[j] = 0
    return list(filter(lambda x: x>0, sieve))
get_prime(10000)
````
