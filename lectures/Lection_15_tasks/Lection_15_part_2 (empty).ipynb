Входные данные:
- массив длиной n элементов.

Алгоритм:
- повторяем n-1 раз;
    - обход всего массива до отсортированной части;
        - если текущий элемент больше следующего, то меняем местами.

````python
import schemdraw
%config InlineBackend.figure_format = 'svg'

schemdraw.theme("dark")

with schemdraw.Drawing() as d:
    d.config(fontsize=11)
    d += (b := flow.Start().label('Начало'))
    d += flow.Arrow().down(d.unit/2)
    d += (d1 := flow.Decision(w=5, h=3.9, W='Да', S='Нет').label('Массив отсортирован n раз?'))
    d += flow.Arrow().length(d.unit/2)
    d += (d2 := flow.RoundBox(5).label("Сортируем каждую пару в\n массиве по очереди"))
    d += flow.Line().down(d.unit/4)
    d += flow.Wire('c', k=3.5, arrow='->').to(d1.E)
    d += flow.Arrow().left().length(2).at(d1.W)
    d += (d3 := flow.StateEnd())
````
